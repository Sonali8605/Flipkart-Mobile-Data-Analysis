- **Sales Conversion Rate** (CR%) Average: 29.75%, meaning around 30 out of 100 visitors buy the phone Best-selling phone has 550% CR 

```python
cat_col=df.select_dtypes(include=['object']).columns
print("categorical column are ")
print(cat_col)
num_col=df.select_dtypes(include=np.number).columns
print("numerical column are ")
print(num_col)
```

## EDA Univarite Analysis (Data Viz)

```python
for i in num_col:
    print(i)
    print("Skeweness:",round(df[i].skew(),2))
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    df[i].hist(grid=True,bins=30)
    plt.ylabel("count")
    plt.subplot(1,2,2)
    sns.boxplot(x=df[i],vert=False)
    plt.show()
    
    
```

```python
for col in num_col:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    df=df[(df[col]>=lower_bound) & (df[col]<=upper_bound)]
df
```

```python
plt.figure(figsize= (18,60))
for i in range(0,5):
    plt.subplot(7,2,i+1)
    plt.title(cat_col[i])
    sns.countplot(df[cat_col[i]])
    
    
plt.show()
```
